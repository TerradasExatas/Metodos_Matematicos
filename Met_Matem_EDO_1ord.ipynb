```python
# EDOs de 1ª ordem
# https://sam-dolan.staff.shef.ac.uk/mas212/notebooks/ODE_Example.html
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

k0=1/5
C1=2
def ydot(y, t):
    ydot=-k0*y+C1
    return ydot

plt.figure()
# solução numérica
t = np.linspace(0,30,10)
y0 = 0.0 
y = odeint(ydot, y0, t)
y = y.flatten()
# y= np.array(y).flatten()

# solução exata
y_exata=y0*np.exp(-k0*t)+(C1/k0)*(1-np.exp(-k0*t))
plt.plot(t,y_exata,'r-')
print(y)
plt.rcParams.update({'font.size': 14})
plt.xlabel("tempo")
plt.ylabel("y")
plt.plot(t, y,'b--');
plt.legend(["solução exata","solução numérica"])
print("size_y=",np.size(y))
#calculando o erro
erro=y-y_exata;
plt.figure()
plt.plot(t,erro,'k')
plt.xlabel("tempo")
plt.ylabel("$y-y_{exata}$")
plt.legend(["erro numérido"])
```
